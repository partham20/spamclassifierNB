```python
import numpy as np 
import pandas as pd 
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.model_selection import train_test_split
```

### GaussianNB Model

```python
#Read Data
data = pd.read_csv('emailsdata.csv')
data
```

```python
x = data.iloc[:, 1:3001]
y = data.iloc[:, -1]
```

```python
#Split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=23)
```

```python
#Create Model
model_1 = GaussianNB()
```

```python
#Fit Model
model_1.fit(x_train, y_train)
```

```python
y_pred_train = model_1.predict(x_train)
y_pred_test = model_1.predict(x_test)
```

```python
#Print Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Train Metrics
print("Gaussian Naive Bayes Classifier - Training Data Metrics:")
print("Train Accuracy:", accuracy_score(y_train, y_pred_train))
print("Train Precision:", precision_score(y_train, y_pred_train))
print("Train Recall:", recall_score(y_train, y_pred_train))
print("Train F1 Score:", f1_score(y_train, y_pred_train))

# Test Metrics
print("\nGaussian Naive Bayes Classifier - Testing Data Metrics:")
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
print("Test Precision:", precision_score(y_test, y_pred_test))
print("Test Recall:", recall_score(y_test, y_pred_test))
print("Test F1 Score:", f1_score(y_test, y_pred_test))
```

# Multinomial Model

```python
#Create Model
model_2 = MultinomialNB()
```

```python
#Fit Model
model_2.fit(x_train, y_train)
```

```python
y_pred_train_2 = model_2.predict(x_train)
y_pred_test_2 = model_2.predict(x_test)
```

```python
#Print Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Train Metrics
print("Multinomial Naive Bayes Classifier - Training Data Metrics:")
print("Train Accuracy:", accuracy_score(y_train, y_pred_train_2))
print("Train Precision:", precision_score(y_train, y_pred_train_2))
print("Train Recall:", recall_score(y_train, y_pred_train_2))
print("Train F1 Score:", f1_score(y_train, y_pred_train_2))

# Test Metrics
print("\nMultinomial Naive Bayes Classifier - Testing Data Metrics:")
print("Test Accuracy:", accuracy_score(y_test, y_pred_test_2))
print("Test Precision:", precision_score(y_test, y_pred_test_2))
print("Test Recall:", recall_score(y_test, y_pred_test_2))
print("Test F1 Score:", f1_score(y_test, y_pred_test_2))
```
